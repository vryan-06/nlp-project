```python
import os
import easyocr
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image_dir = "./images"

image_extensions = {".jpg", ".jpeg", ".png"}
```

```python
image_list = []

for filename in os.listdir(image_dir):
    if any(filename.endswith(ext) for ext in image_extensions):
        image_path = os.path.join(image_dir, filename)
        image = cv2.imread(image_path)
        image_list.append(image)
```

```python
image_list
```

```python
fig = plt.figure(figsize=(16,10))
rows = 3
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(image_list[0])
plt.axis('off')
plt.title("First Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(image_list[1])
plt.axis('off')
plt.title("Second Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(image_list[2])
plt.axis('off')
plt.title("Third Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(image_list[3])
plt.axis('off')
plt.title("Fourth Image");

fig.add_subplot(rows, columns, 5)
plt.imshow(image_list[4])
plt.axis('off')
plt.title("Fifth Image");

fig.add_subplot(rows, columns, 6)
plt.imshow(image_list[5])
plt.axis('off')
plt.title("Sixth Image");

fig.add_subplot(rows, columns, 7)
plt.imshow(image_list[6])
plt.axis('off')
plt.title("Seventh Image");
```

### EASYOCR

```python
text_reader = easyocr.Reader(['en'])
```

```python
output=[]

for im in image_list:
    print("---------------------------")
    t = []
    results = text_reader.readtext(im)
    for (bbox, text, prob) in results:
        print(text)
        t.append(text)
    output.append(t)
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxx")
#     plt.imshow(im)
```
